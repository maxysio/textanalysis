```python
import json
import os
from os.path import join
from ibm_watson import PersonalityInsightsV3
import csv
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```python
# Authentication via IAM

#authenticator = "USE HERE KEY FOR AUTHENTICATION"
service = PersonalityInsightsV3(version='2017-10-13',authenticator=authenticator)
service.set_service_url('https://gateway.watsonplatform.net/personality-insights/api')
```

```python
profile = service.profile('Mary had a little lamb and she love it a lot. She took it everywhere she could.'
                          'She often wondered what she would do if she did not have the little lamb.'
                          'She had really grown attached to it. Now she had to go to school and didnt know'
                          'what to do with the lamb. Then an idea came to her. Maybe she could take the lamb'
                          'to her school. If the teacher catches her she would just take her lamb and come back home.'
                          'Feeling happy that she would spend a lot of time with her lamb, Mary went to sleep.', 
                          'application/json', raw_scores=True,
                          consumption_preferences=True).get_result()

print(json.dumps(profile, indent=2))
```
